```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import arange
import math
```

```python
def quad(x):
    return x**2

def linear(x):
    return x

def exp(x):
    return 2.5**x

def const(x):
    return 1
```

```python
x = arange(0.1,3.1,.1)
plt.title('Comparativo de funções')
plt.xlabel('n')
plt.ylabel('y')

y1 = quad(x)
plt.plot(x,y1)

y2 = linear(x)
plt.plot(x,y2)

y3 = exp(x)
plt.plot(x,y3)

y4 = np.log10(x)
plt.plot(x,y4)

y5 = x*np.log10(x)
plt.plot(x,y5)

y6 = x*const(x)
plt.plot(x,y5)

plt.legend(['y=n^2', "y=n", "y=2.5^n", "y=log(n)", "y=nlog(n)", "y=1"])

plt.savefig('funcoes.pdf')
plt.show()
plt.close()
```
